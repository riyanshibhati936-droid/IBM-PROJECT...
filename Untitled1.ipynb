```python
# Number Guessing Game

import random
secret = random.randint(1, 10)

while True:
  try:
    guess = int(input("Guess the number (1-10):"))
    if guess == secret:
      print("You got it right!")
      break
    elif guess < secret:
      print("Too low! Try again.")
    else:
      print("Too high! Try again.")
  except ValueError:
    print("Invalid input. Please enter a number.")
```
